# Lecture 1

Generally speaking, we've seen that most organizations that think they don't have enough data, actually mean they don't have enough labeled data. If any organization is interested in doing something in practice with a model, then presumably they have some inputs they plan to run their model against. And presumably they've been doing that some other way for a while (e.g. manually, or with some heuristic program), so they have data from those processes! For instance, a radiology practice will almost certainly have an archive of medical scans (since they need to be able to check how their patients are progressing over time), but those scans may not have structured labels containing a list of diagnoses or interventions (since radiologists generally create free text natural language reports, not structured data). We'll be discussing labeling approaches a lot in this book, since it's such an important issue in practice.

Since these kinds of machine learning models can only make predictions (i.e. attempt to replicate labels), this can result in a significant gap between organizational goals and model capabilities. For instance, in this book 


...you'll learn how to create a recommendation system that can predict what products a user might purchase. This is often used in e-commerce, such as to customize products shown on a home page, by showing the highest-ranked items. But such a model is generally created by looking at a user and their buying history (inputs) and what they went on to buy or look at (labels), which means that the model is likely to tell you about products they already have, or already know about, rather than new products that they are most likely to be interested in hearing about. That's very different to what, say, an expert at your local bookseller might do, where they ask questions to figure out your taste, and then tell you about authors or series that you've never heard of before.

1. What is a validation set? What is a test set? Why do we need them?

The `validation` set is not used to update the weights, but it is used to measure model performance on unseed data. By measuring the model accuracy on the `validation` set, we can see when the model starts overfitting on the training set (validation set accuracy will drop). The `validation` set is also used for tuning hyperparameters. Therefore it is important to recognize that this introduces a bias for the model to perform better on the `validation` set than on the `test` set. The `test` set is a part of the dataset that is neither used for training or hyper-parameter tuning. The model performance on the `test` is equivalent to how the model will perform in the real world, assuming that the data generating process has not changed since the `test` set was generated. 


1. What will fastai do if you don't provide a validation set?

It will automatically pick 20% or the dataset for validation. 

1. Can we always use a random sample for a validation set? Why or why not?

When training on a problem where the sequence of examples doesn't matter, then yes, we can random sample the `validation` set. However, when the sequence does matter, e.g., in the case of time series prediction, we cannot do that. When we want to teach the model to extrapolate, it makes no sense to pick the data at random, as we woud teach the model to interpolate. Instead, we need to cut off the last part of the time series and use that as the `validation` set.

## State of Deep Learning



### Image
#### Typical tasks 
- image recognition - classifying an image based on a label (cat og dog)
- object detection - returning a mask showing where in the image an object is (each pixel is apple or not apple)
- image segmentation - returning N masking showing where each N segments are (road, sky, pavement, building, etc.)

#### *out of domain* data 
Deep learning generally performs badly on images in a different style than training data (e.g. training on black and white photos while testing on color images). Therefore it is important to include in the training data all styles expected to appear in production. However, there is no automatic way to check if a data set contains all styles. Data which is not similar in style to any training data is *out of domain*

#### Making more training data with *data augmentation*
Data usually has to be hand labeled/segmented, which is expensive. 
*data augmentation* is used to create variations of the training data and thereby increasing the training set. 

### Text
- Good at machine translation, classification (e.g. sentiment)
- Good at generating context aware text that *sounds* really human and convincig, but not good at generating data that contains meaning or factual information. (I think it's because there's no actual intent behind the answering.)
- danger of text generation: fake news, fake hate messages on a massive scale.
- models for detecting fake text are usually lagging behind models for generation

#### Tabular
- good at Tabular data with high cardinality (zip-code, product id, etc)

#### Reccommendation systems
- Deep learning focussed on making recommendations, not necessarily useful recommendations. It tends to just recommend more of the same. In contrast to a record store guy who knows that music is actually good, and takes your preferences into account while maybe trying to challenge your taste a bit. 

#### Multi-model (combining different kinds of data like images and text)
- Pretty good at things like image captioning, although not necessarily super accurate ("this picture contains two biards", when in fact it contained three birds)
